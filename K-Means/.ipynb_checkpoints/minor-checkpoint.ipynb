# Image to pixels

```python
from PIL import Image
```

```python
im = Image.open("new.png", 'r')
print(im.size)
print(im.mode) #Palette mode, so converting to RGB
```

```python
new_im = im.convert('RGB')
print(new_im.mode)
```

```python
pix_val = list(new_im.getdata())
print(pix_val) #Returning list of tuples
```

```python
# Writing pixel values to file
```

```python
file = open('red.txt','w')
for element in pix_val:
     file.write(str(element[0]))
     file.write(' ')
file.close()
```

```python
file = open('green.txt','w')
for element in pix_val:
     file.write(str(element[1]))
     file.write(' ')
file.close()
```

```python
file = open('blue.txt','w')
for element in pix_val:
     file.write(str(element[2]))
     file.write(' ')
file.close()
```

# Opposite process

```python
file = open('clusters.txt', 'r')

lines = file.readlines()

print(f'Total pixels: {len(lines)}')
```

```python
#Making a dictionary of different clusters
from collections import defaultdict

cluster_pixels = defaultdict(list)

# {
#     0 : [(r, g, b), (r, g, b), ........],
#     1 : [(r, g, b), (r, g, b), ........], 
#     2 : [(r, g, b), (r, g, b), ........]       
# }

for line in lines:
    values = line.split()
    red, green, blue = values[0], values[1], values[2]
    if values[3] == '0':
        cluster_pixels[0].append((int(red), int(green), int(blue)))
    elif values[3] == '1':
        cluster_pixels[1].append((int(red), int(green), int(blue)))
    elif values[3] == '2':
        cluster_pixels[2].append((int(red), int(green), int(blue)))
    else:
        print("Error: Not grouped to any cluster!")
```

```python
len_c0, len_c1, len_c2 = len(cluster_pixels[0]), len(cluster_pixels[1]), len(cluster_pixels[2])
print(len_c0, len_c1, len_c2)
```

```python
#For cluster-0

from PIL import Image
import math

im = Image.new("RGB", (int(math.sqrt(len_c0)), int(math.sqrt(len_c0))))

pixels = im.load()

c0_pixels = cluster_pixels[0]
p = 0 # For maintaining count of pixels

for i in range(int(math.sqrt(len_c0))):
    for j in range(int(math.sqrt(len_c0))):
        pixels[i, j] = c0_pixels[p]
        p += 1

im.show()
im = im.save("first.png")
```

```python
#For cluster-1

from PIL import Image
import math

im = Image.new("RGB", (int(math.sqrt(len_c1)), int(math.sqrt(len_c1))))

pixels = im.load()

c1_pixels = cluster_pixels[1]
p = 0 # For maintaining count of pixels

for i in range(int(math.sqrt(len_c1))):
    for j in range(int(math.sqrt(len_c1))):
        pixels[i, j] = c1_pixels[p]
        p += 1

im.show()
im = im.save("second.png")
```

```python
#For cluster-2

from PIL import Image
import math

im = Image.new("RGB", (int(math.sqrt(len_c2)), int(math.sqrt(len_c2))))

pixels = im.load()

c2_pixels = cluster_pixels[2]
p = 0 # For maintaining count of pixels

for i in range(int(math.sqrt(len_c2))):
    for j in range(int(math.sqrt(len_c2))):
        pixels[i, j] = c2_pixels[p]
        p += 1

im.show()
im = im.save("third.png")
```
